```python
def combineAllSynLocations(aSet):
    mySet = aSet
    AllSyn = {}
    count = 0
    for i in mySet:
        for z in i.synLocation:
            x = z
            while x in AllSyn:
                x = str(z)
                x += str(count)
                x = int(x)
                count +=1
            AllSyn[x] = i.synLocation[z]
    return AllSyn

mySyns = combineAllSynLocations(mySet)
print(mySyns)
#keep a counter of doubles and readd them in when exporting
```

```python
import numpy as np

def cart2sph(x,y,z):
    azimuth = np.arctan2(y,x)
    elevation = np.arctan2(z,np.sqrt(x**2 + y**2))
    r = np.sqrt(x**2 + y**2 + z**2)
    return azimuth, elevation, r

myPoint = cart2sph(365279.5, 301321.6, 229420.4)
print(myPoint)
```
